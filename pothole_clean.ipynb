```python
import pandas as pd
import os
import numpy as np

csv_path = pd.read_csv("pothole_data.csv")
df = pd.DataFrame(csv_path)
df.head()


```

```python
list(df.columns)
```

```python
# Replace col name spaces w/ underscores

df.columns = df.columns.str.replace(' ','_')
list(df.columns)
```

```python
len(df)
```

```python
#Find columns with null values

df.isna().sum()
```

```python
#Values in status
df['STATUS'].unique()

```

```python
#Grab the columns I Want...only remove SSA

df = df.loc[:,['CREATION_DATE',
 'STATUS',
 'COMPLETION_DATE',
 'SERVICE_REQUEST_NUMBER',
 'TYPE_OF_SERVICE_REQUEST',
 'CURRENT_ACTIVITY',
 'MOST_RECENT_ACTION',
 'NUMBER_OF_POTHOLES_FILLED_ON_BLOCK',
 'STREET_ADDRESS',
 'ZIP',
 'X_COORDINATE',
 'Y_COORDINATE',
 'Ward',
 'Police_District',
 'Community_Area',
 'LATITUDE',
 'LONGITUDE',
 'LOCATION']]
```

```python
#count of nulls in zip code
df['ZIP'].isna().sum()
```

```python
#Drop nulls from zip code

df = df[df['ZIP'].notna()]

df['ZIP'].isna().sum()
```

```python
#Find number of null in x_coords and remove those rows
df['X_COORDINATE'].isna().sum()
df = df[df['X_COORDINATE'].notna()]
```

```python
df_local = df[df['LOCATION'].notna()]

df_local.isna().sum()
```

```python
len(df_local)
```

```python
# Looking for null values in status as complete
# condition with df.values property
comp = df_local['STATUS'].values == 'Completed'
  
# new dataframe
new_df = df_local[comp]

new_df.head(10)
```

```python
df_local['Ward'].unique()
```

```python
df_local['Police_District'].unique()
```

```python
df_local['Community_Area'].unique()
```

```python
# Change Zip to String


zip_str = df_local.ZIP.dropna().astype(int).astype(str).str.zfill(5)
zip_str = pd.DataFrame(df_local)
# zip_str.reset_index()
zip_str.head()
```

```python
len(zip_str)
```

```python
# Change Ward to String
w_str = zip_str['Ward'] = zip_str.Ward.dropna().astype(int).astype(str).str.zfill(2)
w_str = pd.DataFrame(zip_str)
w_str.head(10)
```

```python
# Change Police District to String
dist_str = w_str.Police_District.dropna().astype(int).astype(str).str.zfill(2)
dist_str = pd.DataFrame(w_str)
dist_str.head(10)
```

```python
len(dist_str)
```

```python
# Change Community Area to String
ca_str = w_str['Community_Area'] = w_str.Community_Area.dropna().astype(int).astype(str).str.zfill(2)
ca_str = pd.DataFrame(dist_str)
ca_str.head(10)
```

```python
# Reset index for final data frame

reset_df = ca_str.reset_index(drop=True)

reset_df.head()
```

```python
#Sort by creation date

reset_df["CREATION_DATE"] = pd.to_datetime(reset_df["CREATION_DATE"])

sort_df = reset_df.sort_values(by="CREATION_DATE")

sort_df.head()
```

```python
# Create final data frame and reset index

final_df = sort_df.reset_index(drop=True)

final_df.head()
```

```python
#Export to csv to review

final_df.to_csv("potholes_clean.csv")
```
